```python
import pandas as pd
```

```python
def read_dvsa1202(sheets_name):
    df_raw = pd.read_excel(r"C:\Users\cheny\OneDrive - Durham University\其他\UK Driving\dvsa1202.xlsx", sheet_name=sheets_name, header=None)
    df_raw.drop(df_raw.head(6).index,inplace=True)
    df_raw.columns=["center_name","male_take","male_pass","male_r","male_zero","空1","female_take","female_pass","female_r","female_zero","空2","take","pass","r","zero"]
    df=  df_raw.drop(['空1', '空2'], axis=1)
    return df
```

```python
merge_file = []
for i in ["2021-22", "2020-21", "2019-20", "2018-19","2017-18","2016-17","2015-16","2014-15","2013-14","2012-13","2011-12","2010-11","2009-10","2008-09","2007-08"]:
    table = read_dvsa1202(i)
    merge_file.append(table) 
    
result = pd.concat(merge_file)
```

```python
result 
```

```python
result['male_take'] = pd.to_numeric(result['male_take'], errors='coerce')
result['female_take'] = pd.to_numeric(result['female_take'], errors='coerce')
result['female_pass'] = pd.to_numeric(result['female_pass'], errors='coerce')
result['male_pass'] = pd.to_numeric(result['male_pass'], errors='coerce')
result['male_zero'] = pd.to_numeric(result['male_zero'], errors='coerce')
result['female_zero'] = pd.to_numeric(result['female_zero'], errors='coerce')
result['pass'] = pd.to_numeric(result['pass'], errors='coerce')
result['take'] = pd.to_numeric(result['take'], errors='coerce')
result['zero'] = pd.to_numeric(result['zero'], errors='coerce')
```

```python
result =result.groupby('center_name').sum()
```
