```python
## speedytests webpage, for test center ZIP codes
## https://speedytests.co.uk/test-centres.html
import re
import pandas as pd
import urllib.request
import pandas as pd
from fake_useragent import UserAgent
ua=UserAgent()


def get_info(url):
    
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.119 Safari/537.36',
    }
    request = urllib.request.Request(url=url, headers=headers)
    content = urllib.request.urlopen(request).read().decode('utf8')
    return content

```

```python
content = get_info(r"https://speedytests.co.uk/test-centres.html")
```

```python
pattern1 = re.compile(r'col-xs-12 col-md-4"><a href="(.*?)html">')
center_link = pattern1.findall(content)
print(center_link)
```

```python
zip_dict = {}

for i in center_link:
    url_center = i + "html"
    zip_content = get_info(url_center)
    try:
        pattern2 = re.compile(r'<h1>(.*?) Driving Test Centre</h1>')
        center_name = pattern2.findall(zip_content)[0]
        pattern3 = re.compile(r'\b[A-Z]{1,2}[0-9][A-Z0-9]? [0-9][ABD-HJLNP-UW-Z]{2}\b')
        center_zip = pattern3.findall(zip_content)[0]
        zip_dict[center_name] = center_zip
    except:
        zip_dict[" "] = " "
```

```python
pd.DataFrame.from_dict(zip_dict, orient='index').to_csv("center_zip.csv")
print("i")
```
